```python
4. Merge two  sorted linklist into one ?
1->3->5->6->null and 2->4->6->8->null should be merged to make
1->2->3->4->5->6->7->8
```

```python
# step 1 creating the class of linklist 
class ListNode:
    # step 2 creating the constructor  
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
# step 3 creating the founction to merge two linklist 
def mergeTwoLists(l1, l2):
    dummy = ListNode()
    current = dummy
    # step 3 runing the while loop for linklist 1 and the linklist 2
    while l1 and l2:
        if l1.val < l2.val:
            current.next = l1
            l1 = l1.next
        else:
            current.next = l2
            l2 = l2.next
        current = current.next
    
    #step 4  Append remaining elements from either list
    if l1:
        current.next = l1
    else:
        current.next = l2
    
    return dummy.next

#step 5  Example linked lists
list1 = ListNode(1)
list1.next = ListNode(3)
list1.next.next = ListNode(5)
list1.next.next.next = ListNode(6)

list2 = ListNode(2)
list2.next = ListNode(4)
list2.next.next = ListNode(6)
list2.next.next.next = ListNode(8)

# step 6 Merge the two lists
merged_list = mergeTwoLists(list1, list2)

# stp 7  Print the merged list
while merged_list:
    print(merged_list.val, end=" -> ")
    merged_list = merged_list.next
print("null")

```

```python
7.Find the intersection of the two linked lists?
1->2->3->4->8->6->9  5->1->6->7  , intersection 1->6
```

```python
# step 1 creating the class for linklist 
class ListNode:
    # step 2 creating the  constructor for linklist 
    def __init__(self, value=0, next=None):
        self.value = value
        self.next = next
# step 3 calculating the length  of linklist 
def get_length(head):
    length = 0
    current = head
    while current:
        length += 1
        current = current.next
    return length
# step 4  finding the intersection of linklist 
def get_intersection_node(headA, headB):
    lengthA = get_length(headA)
    lengthB = get_length(headB)
    
    # step 5  Align both heads
    while lengthA > lengthB:
        headA = headA.next
        lengthA -= 1
    
    while lengthB > lengthA:
        headB = headB.next
        lengthB -= 1
    
    # step 6 Find intersection
    while headA and headB:
        if headA == headB:
            return headA
        headA = headA.next
        headB = headB.next
    
    return None

# Example usage:
# Constructing the example linked lists
# List 1: 1 -> 2 -> 3 -> 4 -> 8 -> 6 -> 9
# List 2: 5 -> 1 -> 6 -> 7

# Constructing nodes
node1_1 = ListNode(1)
node1_2 = ListNode(2)
node1_3 = ListNode(3)
node1_4 = ListNode(4)
node1_8 = ListNode(8)
node1_6 = ListNode(6)
node1_9 = ListNode(9)

node2_5 = ListNode(5)
node2_1 = ListNode(1)
node2_6 = ListNode(6)
node2_7 = ListNode(7)

# Linking nodes for list 1
node1_1.next = node1_2
node1_2.next = node1_3
node1_3.next = node1_4
node1_4.next = node1_8
node1_8.next = node1_6
node1_6.next = node1_9

# Linking nodes for list 2
node2_5.next = node2_1
node2_1.next = node2_6
node2_6.next = node2_7

# Note: There's no intersection in this specific setup based on references.
intersection = get_intersection_node(node1_1, node2_5)
if intersection:
    print(f"The intersection node's value is {intersection.value}")
else:
    print("No intersection")
```

```python
8. Rotate a linklist by  k  position  to right?
 1->2->3->4->8->6->9 , after rotating for 2 times becomes , 3->4->8->6->9->1->2
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def rotateRight(head: ListNode, k: int) -> ListNode:
    if not head or not head.next or k == 0:
        return head
    
    # Step 1: Calculate the length of the linked list
    length = 1
    old_tail = head
    while old_tail.next:
        old_tail = old_tail.next
        length += 1
    
    # Step 2: Adjust k if necessary
    k = k % length
    if k == 0:
        return head
    
    # Step 3: Find the new head and new tail
    new_tail = head
    for _ in range(length - k - 1):
        new_tail = new_tail.next
    new_head = new_tail.next
    
    # Step 4: Rearrange the links
    new_tail.next = None
    old_tail.next = head
    
    # Step 5: Update the head
    return new_head

# Helper function to print the linked list
def print_list(head):
    while head:
        print(head.val, end=" -> " if head.next else "\n")
        head = head.next

# Example usage
head = ListNode(1)
head.next = ListNode(2)
head.next.next = ListNode(3)
head.next.next.next = ListNode(4)
head.next.next.next.next = ListNode(8)
head.next.next.next.next.next = ListNode(6)
head.next.next.next.next.next.next = ListNode(9)

print("Original list:")
print_list(head)

rotated_head = rotateRight(head, 2)

print("Rotated list by 2 positions to the right:")
print_list(rotated_head)
```

```python
9.Add two number represnted by  linklist
Given two non-empty linked lists representing two non-negative integers, where the digits are stored in
reverse order, add the two numbers and return it as a linked list.
```

```python
# step 1 creating the class of the linklist 
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next
# step 2 creating the 
class Solution:
    def addTwoNumbers(self, l1: ListNode, l2: ListNode) -> ListNode:
        dummy = cur = ListNode(0)  # Create a dummy head node for the result list
        carry = 0

        while l1 or l2 or carry:
            val1 = l1.val if l1 else 0  # Handle cases where one list is shorter
            val2 = l2.val if l2 else 0
            sum = val1 + val2 + carry
            carry, val = divmod(sum, 10)  # Separate carry and digit

            cur.next = ListNode(val)
            cur = cur.next
            l1 = l1.next if l1 else None
            l2 = l2.next if l2 else None

        return dummy.next  # Return the actual result list (excluding the dummy head)

# Example Usage
l1 = ListNode(2, ListNode(4, ListNode(3)))
l2 = ListNode(5, ListNode(6, ListNode(4)))
solution = Solution()
result = solution.addTwoNumbers(l1, l2)

while result:
  print(result.val, end=" -> ")
  result = result.next
print("None")
```

```python
10. Clone a Linked List with next and Random Pointer
Given a linked list of size N where each node has two links: one pointer points to the next node and the
second pointer points to any node in the list. The task is to create a clone of this linked list in O(N) time. 


Note: The pointer pointing to the next node is ‘next‘ pointer and the one pointing to an arCitrary node is
called ‘arbit’ pointer as it can point to any arCitrary node in the linked list. 
```

```python
class Node:
    def __init__(self, val):
        self.val = val
        self.next = None
        self.random = None

def cloneLinkedList(head):
    if not head:
        return None
    
    # Step 1: Create clone nodes and insert them next to original nodes
    current = head
    while current:
        clone_node = Node(current.val)
        clone_node.next = current.next
        current.next = clone_node
        current = clone_node.next
    
    # Step 2: Assign random pointers for clone nodes
    current = head
    while current:
        if current.random:
            current.next.random = current.random.next
        current = current.next.next
    
    # Step 3: Separate the original and cloned lists
    original_head = head
    clone_head = head.next
    cloned_head = clone_head
    while original_head:
        original_head.next = original_head.next.next
        original_head = original_head.next
        if clone_head.next:
            clone_head.next = clone_head.next.next
            clone_head = clone_head.next
    
    return cloned_head
```
